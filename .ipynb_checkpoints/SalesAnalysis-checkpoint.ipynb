### Grouped by Day for October, November, and December

```python
# Extract Day column from Date for each month
df['Day'] = df['Date'].dt.day
df_grouped_Month_Day = df.groupby(['Month', 'Day'])[['Unit', 'Sales']].agg(['mean', 'median', 'max', 'min', 'sum'])
df_grouped_Month_Day.shape
```

```python
def plot_sales_trend_by_day_for_states(month, month_name, states=None):
    # If states are not specified, plot without filtering by state
    if states is None:
        df_grouped_month = df[df['Month'] == month].groupby('Day')['Sales'].sum().reset_index()
        plt.figure(figsize=(10, 4))
        sns.lineplot(x='Day', y='Sales', data=df_grouped_month, marker='o')
        plt.title(f'Total Sales Trend by Day in {month_name} (All States)')
        plt.xlabel('Day')
        plt.ylabel('Total Sales')
        plt.xticks(range(1, 32))  # Setting ticks for each day
        plt.grid(True)
        plt.show()
    else:
        # Calculate grid size: 3 columns per row
        n_states = len(states)
        n_cols = 3
        n_rows = math.ceil(n_states / n_cols)
        
        # Set up the subplot grid
        fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 5 * n_rows), sharey=True)
        axes = axes.flatten()  # Flatten to easily iterate
        
        # Plot each state in its subplot
        for i, state in enumerate(states):
            df_grouped_state_month = (
                df[(df['Month'] == month) & (df['State'] == state)]
                .groupby('Day')['Sales']
                .sum()
                .reset_index()
            )
            sns.lineplot(ax=axes[i], x='Day', y='Sales', data=df_grouped_state_month, marker='o')
            axes[i].set_title(f'{state} - {month_name}')
            axes[i].set_xlabel('Day')
            axes[i].set_ylabel('Total Sales')
            axes[i].set_xticks(range(1, 32))  # Set ticks for each day
            axes[i].grid(True)
        
        # Hide any unused subplots if states are fewer than grid slots
        for j in range(i + 1, len(axes)):
            axes[j].axis('off')
        
        # Adjust layout for readability
        fig.suptitle(f'Total Sales Trend by Day in {month_name} for Selected States')
        plt.tight_layout(rect=[0, 0.03, 1, 0.95])
        plt.show()
```

### Monthly Sales Contribution by Time Across States :

```python
# Sample grouping by State, Month, and Group to get total sales
df_grouped = df.groupby(['State', 'Month', 'Group'])['Sales'].sum().unstack().fillna(0)

# Plotting for each state in separate subplots
states = df['State'].unique()
n_cols = 3  # Adjust the number of columns in the grid layout
n_rows = (len(states) + n_cols - 1) // n_cols  # Calculate rows needed based on state count

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 5), sharey=True)
axes = axes.flatten()  # Flatten for easy indexing

# Loop through each state to create a subplot
for i, state in enumerate(states):
    # Filter data for the specific state
    state_data = df_grouped.loc[state]
    
    # Plot stacked bar for each month with groups stacked in each bar
    state_data.plot(kind='bar', stacked=True, ax=axes[i], colormap='Set3')
    
    # Set title and labels with improved formatting
    axes[i].set_title(f'Monthly Sales Contribution by Group in {state}', fontsize=9, fontweight='bold', color='darkblue')
    axes[i].set_xlabel('Month')
    axes[i].set_ylabel('Total Sales')
    
    # Position legend more neatly
    axes[i].legend(title='Group', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)

# Hide any empty subplots if the number of states is less than grid size
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

# Adjust layout for readability and add a main title
fig.suptitle('Monthly Sales Contribution by Group Across States', fontsize=13, fontweight='bold')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

#### Weekly Sales Contribution by Time Across States :

```python
# Adjust 'Week' column to handle extra days by combining them into the last full week
df['Week'] = df.apply(lambda row: f"{row['Month']}-{min(((row['Day'] - 1) // 7) + 1, 5)}", axis=1)

# Grouping by State, Week, and Time to get total weekly sales per time period within each week and state
df_grouped_weekly_time = df.groupby(['State', 'Week', 'Time'])['Sales'].sum().unstack().fillna(0)

# Define states and set up grid layout for subplots
states = df['State'].unique()
n_cols = 3  # Number of columns for layout
n_rows = (len(states) + n_cols - 1) // n_cols  # Rows needed based on number of states

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 5), sharey=True)
axes = axes.flatten()  # Flatten for easy indexing

# Loop through each state to create a subplot
for i, state in enumerate(states):
    # Filter data for the specific state
    state_data = df_grouped_weekly_time.loc[state]
    
    # Plot stacked bar for each week with time periods stacked in each bar
    state_data.plot(kind='bar', stacked=True, ax=axes[i], colormap='Paired')
    
    # Set title and labels
    axes[i].set_title(f'Weekly Sales Contribution by Time in {state}', fontsize=14, fontweight='bold', color='darkblue')
    axes[i].set_xlabel('Week')
    axes[i].set_ylabel('Total Sales')
    axes[i].set_xticklabels(state_data.index, rotation=45, ha='right')  # Rotate week labels for readability
    
    # Position legend more neatly
    axes[i].legend(title='Time', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)

# Hide any extra subplots if the number of states is less than the grid size
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

# Adjust layout and add a main title
fig.suptitle('Weekly Sales Contribution by Time Across States', fontsize=16, fontweight='bold')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

### 3 - Data Visualization

State-wise sales analysis for different demographic groups Daily

```python
import pandas as pd
import matplotlib.pyplot as plt
import math

# Assuming df has columns 'State', 'Month', 'Day', 'Group', and 'Sales'
# Create a unique identifier for each day across all three months
df['Date'] = df['Month'].astype(str) + '-' + df['Day'].astype(str)

# Define states and set up grid layout for subplots
states = df['State'].unique()
n_cols = 3  # Number of columns in the layout
n_rows = math.ceil(len(states) / n_cols)  # Calculate rows based on number of states

# Set up the subplot grid
fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, n_rows * 5), sharey=True)
axes = axes.flatten()  # Flatten for easy indexing

# Loop through each state and create a plot in a subplot
for i, state in enumerate(states):
    # Filter data for the current state and group by Date and Group to get daily sales
    df_state = df[df['State'] == state]
    df_daily = df_state.groupby(['Date', 'Group'])['Sales'].sum().unstack().fillna(0)
    
    # Normalize each day to 100% to create a percentage contribution
    df_daily_percent = df_daily.div(df_daily.sum(axis=1), axis=0) * 100
    
    # Plot 100% stacked area plot in the appropriate subplot
    df_daily_percent.plot(kind='area', stacked=True, colormap='tab20c', alpha=0.85, ax=axes[i])
    axes[i].set_title(f'{state}', fontsize=14, fontweight='bold')
    axes[i].set_xlabel('Date')
    axes[i].set_ylabel('Percentage of Total Daily Sales (%)')
    axes[i].grid(axis='y', linestyle='--', alpha=0.5)

    # Set x-ticks to show more dates, adjusting as needed
    tick_interval = 5  # Show every 5th day as a tick for better readability
    axes[i].set_xticks(range(0, len(df_daily_percent.index), tick_interval))
    axes[i].set_xticklabels(df_daily_percent.index[::tick_interval], rotation=45)

# Hide any extra subplots if the number of states is less than grid size
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

# Adjust layout and add a main title
fig.suptitle('Daily 100% Stacked Area Plot of Sales Contribution by Group Across States (Oct-Dec)', fontsize=16, fontweight='bold')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import math

# Assuming df has columns 'State', 'Month', 'Group', and 'Sales'

# Define states and set up grid layout for subplots
states = df['State'].unique()
n_cols = 3  # Number of columns in the layout
n_rows = math.ceil(len(states) / n_cols)  # Calculate rows based on the number of states

# Set up the subplot grid
fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, n_rows * 5), sharey=True)
axes = axes.flatten()  # Flatten for easy indexing

# Loop through each state and create a plot in a subplot
for i, state in enumerate(states):
    # Filter data for the current state and create a copy to avoid SettingWithCopyWarning
    df_state = df[df['State'] == state].copy()
    
    # Ensure the Month column is ordered (October = 10, November = 11, December = 12)
    df_state['Month'] = pd.Categorical(df_state['Month'], categories=[10, 11, 12], ordered=True)
    
    # Aggregate data by Month and Group for the current state with observed=False to avoid FutureWarning
    df_monthly = df_state.groupby(['Month', 'Group'], observed=False)['Sales'].sum().unstack().sort_index()

    # Normalize each month to 100% to create a percentage contribution
    df_monthly_percent = df_monthly.div(df_monthly.sum(axis=1), axis=0) * 100
    
    # Plot 100% stacked area plot in the appropriate subplot
    df_monthly_percent.plot(kind='area', stacked=True, colormap='Set3', alpha=0.85, ax=axes[i])
    axes[i].set_title(f'{state}', fontsize=14, fontweight='bold')
    axes[i].set_xlabel('Month')
    axes[i].set_ylabel('Percentage of Total Monthly Sales (%)')
    axes[i].grid(axis='y', linestyle='--', alpha=0.5)

    # Set x-ticks and labels to show October, November, and December
    axes[i].set_xticks([0, 1, 2])
    axes[i].set_xticklabels(['October', 'November', 'December'], rotation=0)

# Remove any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout and add a main title
fig.suptitle('Monthly 100% Stacked Area Plot of Sales Contribution by Group for Each State (Oct-Dec)', fontsize=16, fontweight='bold')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
# Grouping by 'State' and 'Group', calculating mean, and reshaping the data
avg_sales_pivot = df.groupby(['Group', 'State'])['Sales'].mean().round(2).unstack()
# Normalize to get percentages by dividing each cell by the row (state) total
avg_sales_percentage = avg_sales_pivot.div(avg_sales_pivot.sum(axis=1), axis=0)

# Plotting the normalized stacked bar chart
avg_sales_percentage.plot(kind='bar', stacked=True, figsize=(10, 6))

# Adding labels and title
plt.title("Quarterly Average Sales by Group and State (100% Stacked)")
plt.xlabel("Group")
plt.ylabel("Percentage of Sales")
plt.legend(title="State", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

Time of the day analysis 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df has columns 'Month', 'Day', 'Time', and 'Sales'
# Create a unique identifier for each day across all three months
df['Date'] = df['Month'].astype(str) + '-' + df['Day'].astype(str)

# Aggregate data by Date and Time across all states
df_daily_total = df.groupby(['Date', 'Time'])['Sales'].sum().unstack().fillna(0)

# Normalize each day to 100% to create a percentage contribution
df_daily_percent = df_daily_total.div(df_daily_total.sum(axis=1), axis=0) * 100

# Plotting 100% stacked area plot
plt.figure(figsize=(15, 8))
df_daily_percent.plot(kind='area', stacked=True, colormap='tab20c', alpha=0.85)

# Customizing the plot
plt.title('Daily 100% Stacked Area Plot of Sales Contribution by Time (Oct-Dec)')
plt.xlabel('Date')
plt.ylabel('Percentage of Total Daily Sales (%)')
plt.xticks(rotation=45)  # Rotate x-ticks for better readability
plt.legend(title='Time of Day', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
